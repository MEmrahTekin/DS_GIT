#stack Plot
<h1>   Method used to calculate the baseline:

- `'zero'`: Constant zero baseline, i.e. a simple stacked plot.
- `'sym'`:  Symmetric around zero and is sometimes called
  'ThemeRiver'.
- `'wiggle'`: Minimizes the sum of the squared slopes.
- `'weighted_wiggle'`: Does the same but weights to account for
  size of each layer. It is also called 'Streamgraph'-layout. More </h1>

```python
x = ["a","b","c","d","e","f"]
y = random.choices(range(6,10),k=6)
y2 = random.choices(range(1,3),k=6)
y3 = random.choices(range(2,6),k=6)
print(x,y,y2,y3,sep="\n")
```

```python
plt.stackplot(x,y,y2,y3 ,labels=["y","y2","y3"])
plt.title("Stack Chart\nExample")
plt.xlabel("X' axis LABEL")
plt.ylabel("Y' axis LABEL")
plt.legend()
plt.show()
```

```python
plt.stackplot(x,y,y2,y3 ,labels=["y","y2","y3"],baseline="sym")
plt.title("Stack Chart\nExample")
plt.xlabel("X' axis LABEL")
plt.ylabel("Y' axis LABEL")
plt.legend()
plt.show()
```

```python
plt.stackplot(x,y,y2,y3 ,labels=["y","y2","y3"],baseline="weighted_wiggle")
plt.title("Stack Chart\nExample")
plt.xlabel("X' axis LABEL")
plt.ylabel("Y' axis LABEL")
plt.legend()
```

```python
plt.stackplot(x,y,y2,y3 ,labels=["y","y2","y3"],baseline="wiggle")
plt.title("Stack Chart\nExample")
plt.xlabel("X' axis LABEL")
plt.ylabel("Y' axis LABEL")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn( 4,100)
fig, ax = plt.subplots(1, 2)
my_plotter(ax[0], data1, data2, {'marker': '^', "color":"r", "linewidth": 3, "linestyle":"-"} )
my_plotter(ax[1], data1, data2, {'marker': 'o', "color":"b", "linewidth": 1, "linestyle":"--", "markeredgecolor": "r"})
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
